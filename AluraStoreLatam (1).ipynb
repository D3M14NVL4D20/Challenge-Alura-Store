### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
import pandas as pd
```

```python
df = tienda['Categoría del Producto'].value_counts()
df
df2 = tienda2['Categoría del Producto'].value_counts()
df2
df3 = tienda3['Categoría del Producto'].value_counts()
df3
df4 = tienda4['Categoría del Producto'].value_counts()
df4
```

```python
# Lista de DataFrames y nombres
dataframes = [df, df2, df3, df4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i in range(len(dataframes)):
    plt.figure(figsize=(8,5))
    ax = dataframes[i].plot(kind='bar', color='darkcyan', edgecolor='black')
    plt.title(f'Categorías Vendidas - {nombres_tiendas[i]}')
    plt.xlabel('Categorías')
    plt.ylabel('Cantidad de productos vendidos')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    color = colores = ['mediumseagreen', 'goldenrod', 'steelblue', 'plum']
    # Etiquetas encima de cada barra
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom', fontsize=10, color='black')

    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Diccionario con nombres y promedio de calificación
calificaciones_tiendas = {
    'Tienda 1': tienda['Calificación'].mean(),
    'Tienda 2': tienda2['Calificación'].mean(),
    'Tienda 3': tienda3['Calificación'].mean(),
    'Tienda 4': tienda4['Calificación'].mean()
}

print("Calificaciones por tienda:", calificaciones_tiendas)


tienda_mejor_calificada = max(calificaciones_tiendas, key=calificaciones_tiendas.get)
tienda_peor_calificada = min(calificaciones_tiendas, key=calificaciones_tiendas.get)

print("Tienda mejor calificada:", tienda_mejor_calificada)
print("Tienda peor calificada:", tienda_peor_calificada)
```
